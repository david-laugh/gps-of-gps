```python
df = pd.read_csv("art3_USA_CA000000000000")
for i in range(1, 10):
    df1=pd.read_csv('art3_USA_CA00000000000%d'%i)
    df=pd.concat([df,df1])
```

```python
def rst(a):
    a=a.reset_index()
    a=a.drop('index', axis=1)
    return a
```

```python
df=rst(df)
```

```python
lt=list(df['lat'])
ln=list(df['lon'])
```

```python
f=[]
s=[]
for i in tqdm(range(len(lt))):
    f.append(h3.geo_to_h3(lt[i], ln[i], 5))
    s.append(h3.geo_to_h3(lt[i], ln[i], 6))
```

```python
f=pd.DataFrame(f)
s=pd.DataFrame(s)
```

```python
df['h3(Lv5)']=f[0]
df['h3(Lv6)']=s[0]
```
